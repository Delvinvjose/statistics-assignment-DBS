#(d)	Explain how the box-plot technique can be used to detect outliers. Apply this technique for one attribute of the dataset                 

```r
X=dataa$MaxTemp
q=quantile(X)
q1=q[2]
q3=q[4]
iqr=q3-q1
lowerr=q1-1.5*iqr
upperr=q3+1.5*iqr
out=a[a<lowerr|a>upperr]
boxplot(a, main = "Box Plot of MaxTemp", ylab = "MaxTemp",notch = TRUE,outline = TRUE)
# print(out)
cat("Outliers using custom multiplier (2 * IQR):", out, "\n")
```

```r
#Rainfall

X= dataa$Rainfall[dataa$Rainfall > 0]

estimated_lamda = 1 / mean(X)

#simulate the data for next 10 days using exponential distribution
predicted_rainfall = rexp(10,rate = estimated_lamda);predicted_rainfall

x = seq(0, max(predicted_rainfall) * 1.5, length.out = 100)
pdf_vals = dexp(x, rate = estimated_lamda)

# Plot PDF
plot(x, pdf_vals, type = 'l', col = 'blue', main = "Predicted PDF of Rainfall",
     xlab = "Rainfall", ylab = "Density")
```

The generated values in predicted_rainfall represents simulated rainfall prediction for the next 10 days,assuming it follows exponential distribution.

(b) Consider one categorical variable, apply goodness of fit test to evaluate whether a candidate set of probabilities can be appropriate to quantify the uncertainty of class frequency at the significant level alpha=0.05.                 

hypothesis

H0= Categorical variable matches the expected probability

H1= categorical variable doesnot match the expected probability

```r
A=dataa$RainToday
B=as.factor(A)
observed=table(B)
expected=c(No=0.7,Yes=0.3)
total=sum(observed)
expect_fre=expected*total
test=chisq.test(observed,p = expected)
# print(observed)
print(expect_fre)
stat=test$statistic
p_value=test$p.value;p_value
alpha=0.05
if(p_value<alpha){
  cat("reject")
}else{
  cat("accept")
}
```

Since Test value is less than the alpha value ,H0 is rejected.

(c) Consider one continuous variable in the dataset, and apply test of mean for a proposed candidate of  at the significant level alpha=0.05

H0: Sample mean is significantly different
<br>H1:Sample mean is significantly not different

```r
max_temp=dataa$MaxTemp
mu0=25
test=t.test(max_temp,mu = mu0,alternative = "two.sided",conf.level = 0.95)
print(test)
p_value=test$p.value
if(p_value<0.05){
  cat("sample mean is significantly different")
}else{
  cat("sample is not significantly different")
}
```
